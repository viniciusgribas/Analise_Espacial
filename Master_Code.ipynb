```python
# Preservando o gdf original

gdf_br = gdf_brasil_shp.copy()

# Modificando o nome das colunas do DF

gdf_br["NM_ESTADO"] = gdf_br.NM_ESTADO.map({
'RONDÔNIA':'RO', 'ACRE':'AC', 'AMAZONAS':'AM', 'RORAIMA':'RR', 'PARÁ':'PA', 'AMAPÁ':'AP',
       'TOCANTINS':'TO', 'MARANHÃO':'MA', 'PIAUÍ':'PI', 'CEARÁ':'CE', 'RIO GRANDE DO NORTE':'RN',
       'PARAÍBA':'PB', 'PERNAMBUCO':'PE', 'ALAGOAS':'AL', 'SERGIPE':'SE', 'BAHIA':'BA',
       'MINAS GERAIS':'MG', 'ESPIRITO SANTO':'ES', 'RIO DE JANEIRO':'RJ', 'SÃO PAULO':'SP',
       'PARANÁ':'PR', 'SANTA CATARINA':'SC', 'RIO GRANDE DO SUL':'RS',
       'MATO GROSSO DO SUL':'MS', 'MATO GROSSO':'MT', 'GOIÁS':'GO', 'DISTRITO FEDERAL':'DF'})


# Alterando o nome para UF, assim conicidirá com o outro df

gdf_br = gdf_br.rename(columns = {'NM_ESTADO': 'UF'}, inplace = False)

# Criando um Dataframe Geográfico a partir do DF antigo

gdf_ger = gpd.GeoDataFrame(
    df_ger, geometry=gpd.points_from_xy(df_ger.X, df_ger.Y))

# Determinando um CRS para o novo DF Geográfico

gdf_ger.crs = "EPSG:5880" 

# Unindo ambos os Dataframes com o comando ".merge". 
# A coluna 'UF' une o que foi originado do arquivo shp com geometria de polígnos e o que foi originado do arquivo ger com geometria de pontos.

gdf_br = gdf_br.merge(gdf_ger, on='UF')

# Renomeando as novas colunas geometry_x e geometry_y. As bibliotecas por padrão apenas conseguem ler um tipo de geometria, logo, 
# Excluímos a geometria de ponto para esta análise

gdf_br = gdf_br.rename(columns = {'geometry_x': 'geometry','geometry_y': 'geometry_point' }, inplace = False)
#gdf_br = gdf_br.drop('geometry_point', axis='columns')


```

```python
# Agora, busca-se criar um novo DF, menor, contendo observações referentes aos valores agrupados por UF

### Primeiro, se extraí a soma total de cada um dos tipos de geração

gpby_gdf_br_uf_sum = gdf_br.groupby(['UF','TipoGeracao']).sum().reset_index()

### Depois, soma-se a quantidade dos tipos de geração para cada UF. Excluindo as colunas desnecessárias no final
gpby_gdf_br_uf_tipoGeracao_count = gdf_br.groupby(['UF','TipoGeracao']).count().reset_index()

##### Excluindo as colunas desnecessárias no final
gpby_gdf_br_uf_tipoGeracao_count = gpby_gdf_br_uf_tipoGeracao_count[['UF','TipoGeracao','ID']].copy()
##### Renomeando-a para 'Cont'
gpby_gdf_br_uf_tipoGeracao_count.rename(columns={"ID": "Cont"},inplace = True)


### Agora, cria-se um DF com as geometrias, para atribuí-las ao mapa
newdf = gdf_br[['UF','geometry','TipoGeracao']].copy() 
gdf_br_UF_geometry_TipoGera = newdf.drop_duplicates()
```

```python
# Cria-se uma coluna CHAVE para a união dos DataFrames de soma, contagem e geometria. 
# Esta coluna contem valores únicos referentes à Tipo de Geração e UF.

# DataFrame de Soma
gpby_gdf_br_uf_sum['CHAVE'] = gpby_gdf_br_uf_sum['UF']+gpby_gdf_br_uf_sum['TipoGeracao']

# DataFrame de Contagem
gpby_gdf_br_uf_tipoGeracao_count['CHAVE'] = gpby_gdf_br_uf_tipoGeracao_count['UF']+gpby_gdf_br_uf_tipoGeracao_count['TipoGeracao']

# DataFrame de Geometria
gdf_br_UF_geometry_TipoGera['CHAVE'] = gdf_br_UF_geometry_TipoGera['UF']+gdf_br_UF_geometry_TipoGera['TipoGeracao']


# Primeiro unem-se os DataFrames de Soma e Contagem
gdf_br_final_0 = pd.merge(gpby_gdf_br_uf_tipoGeracao_count,gpby_gdf_br_uf_sum ,on= 'CHAVE', how='inner')

# Renomeiam-se as colunas duplicadas que foram atribuídos os sufixos "_X" e "_Y". Mantendo as com o sufixo "_X" e excluíndo as com sufixo "_Y"
gdf_br_final_0.rename(columns={"UF_x": "UF","TipoGeracao_x":"TipoGeracao"},inplace = True)
gdf_br_final_0.drop( columns = {"UF_y","TipoGeracao_y"}, inplace = True)


gdf_br_final = pd.merge(gdf_br_UF_geometry_TipoGera,gdf_br_final_0 ,on= 'CHAVE', how='inner')
gdf_br_final.rename(columns={"UF_x": "UF","TipoGeracao_x":"TipoGeracao"},inplace = True)
gdf_br_final.drop( columns = {"UF_y","TipoGeracao_y"}, inplace = True)



gdf_br_final.drop( columns = {"CHAVE","ID","MdaPotenciaFinalizadaKW"}, inplace = True)

gdf_br_final.rename(columns={"MdaPotenciaOutorgadaKW": "PotenciaOutorgadaKW","MdaGarantiaFisicaKW":"GarantiaFisicaKW"},inplace = True)

gdf_br_final.isnull().sum()



```
