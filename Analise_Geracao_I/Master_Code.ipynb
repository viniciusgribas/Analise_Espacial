### Data Wrangling

```python
# Primeiro, preserva-se o dataframe inicial "df_ini", realizando uma cópia deste.

df = df_ini.copy()

# Aqui se inicia a análise exploratória dos arquivos, a função ".info" permite 
# além de verificar as colunas, ver seu dtype.

print(df.info())
```

```python
# A função abaixo, permite verificar o nome atribuído a cada uma das colunas

print(df.columns.unique())
```

```python
# Realiza-se então a exclusão das que não serão utilizadas para esta análise

df = df.drop(columns=['DatGeracaoConjuntoDados','IdcGeracaoQualificada','CodCEG','DscTipoOutorga','DatEntradaOperacao','MdaPotenciaFinalizadaKw','MdaGarantiaFisicaKw','DatInicioVigencia','DatFimVigencia','DscPropriRegimePariticipacao','DscSubBacia','DscMuninicpios' ] )
df.info()
```

```python
# -  NomEmpreendimento         Variável Qualitativa Nominal   
# -  SigUFPrincipal            Variável Qualitativa Nominal   
# -  DscAbreviacaoTipoGeracao  Variável Qualitativa Nominal   
# -  DscFaseUsina              Variável Qualitativa Ordinária   
# -  DscOrigemCombustivel      Variável Qualitativa Nominal   
# -  DscFonteCombustivel       Variável Qualitativa Nominal   
# -  NomFonteCombustivel       Variável Qualitativa Nominal   
# -   MdaPotenciaOutorgadaKw  Variável Quantitativa Contínua
# -  NumLatitude               Variável Qualitativa Ordinária   
# -  NumLongitude              Variável Qualitativa Ordinária   
```

```python
# Verifica-se as usinas que estão em operação, que estão em construção e 
# que ainda não foram iniciadas.

df.groupby("DscFaseUsina").count()
```

```python
# Filtra-se apenas as que estão em operação

df = df.query("DscFaseUsina == 'Operação'")
df.groupby("DscFaseUsina").count()
```

```python
# Aqui se gera uma tabela por 'DscOrigemCombustivel' com 
# valores totais (nulos e não nulos)


temp_a = df.groupby("DscOrigemCombustivel").sum()

print('============================TABELA DE VALORES TOTAIS====================================')

print(temp_a)



print('========================================================================================')

# Aqui se gera uma tabela por 'DscOrigemCombustivel' com valores nulos

print('============================TABELA DE VALORES NULOS====================================')

null_data = df[df.isnull().any(axis=1)]

temp_b = null_data.groupby("DscOrigemCombustivel").sum()

print(temp_b)

print('========================================================================================')
```

Verifica-se que o percentual dos valores Null, quando comparados com os totais, são muito baixos.

```python
# Aqui se gera uma tabela por 'DscOrigemCombustivel' com a representatividade 
# dos valores nulos

print('============================DIFERENÇA PERCENTUAL DE VALORES NULOS====================================')

dif = (((temp_a-temp_b)/temp_a)*100)

print(100 - dif )
```

```python
# Verifica-se agora, por coluna os valores não nulos, 
# nulos e por fim se excluí os nulos desta análise.

print(df.count())

print('===============')

print(df.isnull().sum())

print('===============')

print(df.dropna().count())

print('XXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXX')

print(100 - dif )

# Novo DF, agora pronto para as análises "df_fin"

df_ger = df.dropna().copy()

df_ger.head()
```

```python
# Modificando o nome das colunas do DF

df_ger = df_ger.rename( columns ={
"NomEmpreendimento":"Empreendimento",
"SigUFPrincipal":"UF",
"DscAbreviacaoTipoGeracao":"TipoGeracao",
"DscFaseUsina":"Fase",
"DscOrigemCombustivel":"OrigemCombustivel",
"DscFonteCombustivel":"FonteCombustivel",
"NomFonteCombustivel":"NomeCombustivel",
"MdaPotenciaOutorgadaKw":"MdaPotenciaOutorgadaKW",
"MdaPotenciaFinalizadaKw":"MdaPotenciaFinalizadaKW",
"MdaGarantiaFisicaKw":"MdaGarantiaFisicaKW",
"NumLongitude": "X",
"NumLatitude": "Y",	
"DscMuninicpios":"Municipio"
})
```

### Análises Exploratórias Geoespaciais I *[Geometria - Ponto]*

```python
# Com os dados minimamente lapidados, agora é possível iniciar as fases de 
# Análises Exporatórias Geoespaciais na geometria ponto

df_ger.head()
```

```python
# Contando quantos empreendimentos existem em operação, para verificar se condiz 
# com o relatório disponibilizado plea ANEEL original (2021)

df_ger.groupby('TipoGeracao').count()
```

```python
title_BarChart = "Diagrama de Barras:Potência Outorgada [KW] e Tipo de Geração por UF"

title_BarChart_export = "Bar"+github


fig_BarChart = px.bar(df_ger, x="UF",
 y="MdaPotenciaOutorgadaKW",
 color="TipoGeracao", 
 title=title_BarChart,
 width=800, 
 height=1000)
fig_BarChart.show()

plt.offline.plot(fig_BarChart, filename = title_BarChart_export)
```

```python
title_strip = "Diagrama de Pontos:Potência Outorgada [KW] e Tipo de Geração por UF"

title_strip_export = "Strip"+github

fig_Strip = px.strip(df_ger, title=title_strip, x="MdaPotenciaOutorgadaKW", y="UF", orientation="h", color="TipoGeracao")
fig_Strip.show()

plt.offline.plot(fig_Strip, filename = title_strip_export)
```

```python
title_Polar = "Diagrama Polar: Potência Outorgada[KW] e UF por Tipo de Geração"
title_Polar_export = "Polar"+github

fig_Polar = px.bar_polar(df_ger, title=title_Polar, r="MdaPotenciaOutorgadaKW", theta="UF", color="MdaPotenciaOutorgadaKW", template="plotly",animation_frame="TipoGeracao",
            color_discrete_sequence= px.colors.sequential.Plasma_r)
fig_Polar.show()

plt.offline.plot(fig_Polar, filename = title_Polar_export)
```

```python
df_ger.TipoGeracao.unique()
```

```python
title_Densidade = "Diagrama Densidade: Potência Outorgada[KW] por Tipo de Geração"
title_Densidade_export = "Densidade"+github

fig_dens = px.density_heatmap(df_ger, title=title_Densidade, x="UF", y="TipoGeracao", marginal_x="histogram", marginal_y="histogram",text_auto=True)
fig_dens.show()
plt.offline.plot(fig_dens, filename = title_Densidade_export)
```

```python
df_ger['X'] = df_ger['X'].apply(lambda col:pd.to_numeric(col, errors='coerce'))
df_ger['Y'] = df_ger['Y'].apply(lambda col:pd.to_numeric(col, errors='coerce'))

df_ger.info()
```

```python
# Criando um Dataframe Geográfico a partir do DF antigo

gdf_ger = gpd.GeoDataFrame(
    df_ger, geometry=gpd.points_from_xy(df_ger.X, df_ger.Y))

# Determinando um CRS para o novo DF Geográfico

gdf_ger.crs = "EPSG:5880" 


title_pointsMap = "Mapa de Coordenadas: Potência Outorgada por Localização do Empreendimento"
title_pointsMap_export = "Points"+github

fig_pointsMap = px.scatter_mapbox(df_ger,
                title=title_pointsMap,
                lat="Y", 
                lon="X", 
                color="MdaPotenciaOutorgadaKW", 
                size="MdaPotenciaOutorgadaKW",
                color_continuous_scale=px.colors.sequential.matter,
                hover_name = "Empreendimento" ,
                hover_data = ["MdaPotenciaOutorgadaKW"], 
                size_max=80, 
                zoom=3,
                mapbox_style='open-street-map', 
                animation_frame='TipoGeracao')
fig_pointsMap.show()
plt.offline.plot(fig_pointsMap, filename = title_pointsMap_export)
```
